```python
from sxtlac import LAC
xx = np.linspace(200, 1000, 100)
lac_H2O = LAC('H2O',1)
yy = [lac_H2O(x) for x in xx]
plt.plot(xx,yy,label = 'H2O')

lac_SiO2 = LAC('SiO2',2.2)
yy = [lac_SiO2(x) for x in xx]
plt.plot(xx,yy,label = 'SiO2')

plt.legend()
```
